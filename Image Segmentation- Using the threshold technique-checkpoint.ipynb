# Region -Based Segmentation
# Implementing thresholding in Image segmentation.

This project aims to look out how region based segmentation works such that objects are segmented using their pixel values.
We then set a threshold value where the pixel values falling below or above that thresholdcan be classified as an object or the background.

This project utilizes a global threshold because the image used will be divided into two regions(object and background), hence
the need to define a single threshold value.

```python
#import required libraries

from skimage.color import rgb2gray
```

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

```python
# read image and plot it
image = plt.imread("C:/Users/adeol/Pictures/20200423_143623.jpg")
image.shape
plt.imshow(image)
```

```python
image.shape #check the shape of the image
```

```python
#The image is a three-channel image(RGB);
#convert it into a single channel to understand the algorithm

gray = rgb2gray(image)
plt.imshow(gray, cmap="gray")
gray.shape
```

We will then apply a certain threshold to this image such that it will seperate the image into two parts i.e
the object and background. Since the height and width of the image is 2322 and 4128 respectively, the mean of the pixel value
will be used as a threshold. The conditions for selection includes:
    - If the pixel value is more than the threshold established, we can say it belongs to an object
    - If the pixel value is less than the threshold established, we can say it belongs to the background

```python
# reshape the grayscale
gray_rs = gray.reshape(gray.shape[0]*gray.shape[1])

for i in range(gray_rs.shape[0]):
    if gray_rs[i] > gray_rs.mean():
        gray_rs[i] = 1
    else:
        gray_rs[i] = 0
gray = gray_rs.reshape(gray.shape[0],gray.shape[1])

plt.imshow(gray, cmap='gray')
```
